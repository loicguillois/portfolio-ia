```python
import spacy

# Charger le modèle CamemBERT
nlp = spacy.load('fr_dep_news_trf')

def get_embedding(text):
    if text is None:
        return None
    doc = nlp(text)
    return doc._.trf_data.tensors[-1][0]

# Obtention des embeddings pour toutes les descriptions
filtered_df['embedding_histoire'] = filtered_df['Histoire'].apply(get_embedding)
```

```python
filtered_df['embedding_atouts'] = filtered_df['Atout'].apply(get_embedding)
```

```python
filtered_df = pd.read_pickle('df.pkl')
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calcul de la matrice de similitude des cosinus

# TODO : filter sur departement

similarity_matrix = cosine_similarity(list(filtered_df['embedding_histoire']))

# Fonction pour obtenir les recommandations
def get_recommendations(index, num_recommendations=5):
    sim_scores = list(enumerate(similarity_matrix[index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:num_recommendations+1]
    museum_indices = [i[0] for i in sim_scores]
    return filtered_df.iloc[museum_indices]

# Obtention des recommandations pour un musée spécifique
recommendations = get_recommendations(673, 5)
print(recommendations)
```

```python
import numpy as np

def get_average_embedding(row):
    atouts_embedding = row['embedding_atouts']
    histoire_embedding = row['embedding_histoire']
    
    if atouts_embedding is None or histoire_embedding is None:
        return None
    
    return (atouts_embedding + histoire_embedding) / 2

filtered_df['average_embedding'] = filtered_df.apply(get_average_embedding, axis=1)

# Suppression des lignes où les embeddings ne sont pas disponibles
filtered_df.dropna(subset=['average_embedding'], inplace=True)

# # Calcul de la matrice de similitude cosinus
X = np.stack(filtered_df['average_embedding'].values)
similarity_matrix = cosine_similarity(X)

recommendations = get_recommendations(673, 5)
print(recommendations)
```
